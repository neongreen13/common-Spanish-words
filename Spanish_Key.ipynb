# Step 1: Download 250 Common Spanish Words as Word Key 

Tools Used:
- Web scrape tables using read_html in Pandas

Starting Page
https://www.happyhourspanish.com/learning-efficiently-start-with-the-250-most-common-spanish-words/

### Download HTML Tables

```python
import pandas as pd

dfs = pd.read_html('https://www.happyhourspanish.com/learning-efficiently-start-with-the-250-most-common-spanish-words/',header=0)
for df in dfs:
    print(df) #loops all 3 tables
```

```python
dfs #displays all 3 tables as 3 separate lists
```

```python
spanish = pd.concat(dfs)
spanish
```

```python
pd.DataFrame(spanish).to_excel('Spanish_Key.xlsx', header=True, index=True)
```

```python
pd.read_excel('Spanish_Key.xlsx',sheet_name='Sheet1')
```

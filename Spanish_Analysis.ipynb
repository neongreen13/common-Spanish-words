### Analysis

Bag of Words Model:

Text must be converted to numbers for machine learning, or feature extraction. The bag-of-words model is used here because it describes the occurrence of words within the lyrics corpus. The word order is unimportant to this analysis, since we simply want to know the most commonly used Spanish words, or word frequency. Our goal is to find the top 200-300 words, so we can learn them first as an efficient way to gain a grasp of the 20% of all Spanish words that are spoken 80% of the time.

##### Compare and contrast Bag of Words lists

```python
import re
def preprocess(data):
    reviews_tokens = []
    stopwords=[]
    for review in data:
        review = review.lower() #Convert to lower-case words
        raw_word_tokens = re.findall(r'(?:\w+)', review,flags = re.UNICODE) #remove pontuaction
        word_tokens = [w for w in raw_word_tokens if not w in stopwords] # do not add stop words
        reviews_tokens.append(word_tokens)
    return reviews_tokens #return all tokens
```

```python
def construct_bag_of_words(data):
    corpus = preprocess(data)
    bag_of_words = {}
    word_count = 0
    for sentence in corpus:
        for word in sentence:
            if word not in bag_of_words: # do not allow repetitions
                bag_of_words[word] = word_count #set indexes
                word_count+=1
            
    return bag_of_words #index of letters
```

```python
bag_of_words_250 = construct_bag_of_words(lyrics_corpus_1['Lyrics']) #made the words into a dictionary
bag_of_words_100 = construct_bag_of_words(lyrics_corpus_100['Lyrics'])
bag_of_spanish_key = construct_bag_of_words(spanish['Spanish'])
bag_of_corp_1 = construct_bag_of_words(text_corp_1['Word form'])
bag_of_corp_2 = construct_bag_of_words(text_corp_2['Lemma'])
```
